```python
from surprise import Dataset
data = Dataset.load_builtin('ml-100k')
#import puis visualisation de données
data.__dict__
```

```python
import pandas as pd
df = pd.DataFrame(data.raw_ratings,columns = ('uid','iid','grade','id'),dtype = int)
```

```python

df.describe()
```

```python
df.head()
```

```python
%matplotlib inline
grades_column = df["grade"]
grades_column.plot(kind="hist" , bins = np.arange(0.5,6.5,1) , xticks = (1,2,3,4,5))
# on ajoute des bins 0.5 9 par pas de 1 et ticks 
```

```python
from surprise import KNNBasic
algo = KNNBasic()
```

```python
#Creation des corpus
from surprise.model_selection import train_test_split
trainset, testset = train_test_split (data, test_size = .20)
```

```python
algo.fit(trainset)
predictions = algo.test(testset)
```

```python
from surprise import accuracy
accuracy.mse(predictions)
```

```python
# plot une fonction avec les accuracy en fonction des k

for k in [2, 10, 70 ,80 , 800, 8000]:
    algo = KNNBasic(k)
    algo.fit(trainset)
    predictions = algo.test(testset)
    print ("k = ", k, accuracy.mse(predictions))
```

```python
#Correction la distance de Manhatan n'est pas efficace
#Distance Eucludient plus efficace sur des valeurs réels
#Distance Cosine est plus efficace sur les valeurs boolean
import mlflow
from surprise import KNNWithMeans
for k in range(20,40):
    for distance in [{'name': 'msd','user-based' : True, 'min_support' : 5}, {'name': 'cosine','user-based' : True, 'min_support' : 5} ] :
        mlflow.start_run()
        mlflow.log_param("k",k)
        mlflow.log_param("distance",distance["name"])
        algo = KNNWithMeans(k=k,sim_options = distance)
        
        start_time = time.time()
        algo.fit(trainset)
        predictions = algo.test(testset)
        
        temps = time.time() - start_time
        
        #then compute MSE
        acc = accuracy.mse(predictions)
        
        mlflow.log_metric=("temps",temps)
        mlflow.log_metric=("MSE",acc)
        mlflow.end_run()
```
